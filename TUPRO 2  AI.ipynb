```python
# Read Data Set
import pandas as pd
import matplotlib.pyplot as plt

vendor = pd.read_excel("supplier.xlsx")
vendor
```

```python
# Inisiasi Data
id = vendor["id"]
kualitas = vendor["kualitas"]
harga = vendor["harga"]
```

```python
y = []

for x in range(100):
#Fuzzyfication
  nilaiKain = [0,0,0,0]
  nilaiHarga = [0,0,0,0]
  SangatJelek = Jelek = Bagus = SangatBagus = Murah = Sedang = Mahal = 0
  if kualitas[x] <= 20:
    SangatJelek = 1
    nilaiKain[0] = SangatJelek
  elif kualitas[x] > 20 and kualitas[x] < 25:
    SangatJelek = (25 - kualitas[x])/5
    Jelek = (kualitas[x] - 20)/5
    nilaiKain[0] = SangatJelek
    nilaiKain[1] = Jelek
  elif kualitas[x] >=25 and kualitas[x] <=45:
    Jelek = 1
    nilaiKain[1] =Jelek
  elif kualitas[x] > 45 and kualitas[x] < 50:
    Jelek = (50 - kualitas[x])/5
    Bagus = (kualitas[x] - 45)/5
    nilaiKain[1] = Jelek
    nilaiKain[2] = Bagus
  elif kualitas[x] >=50 and kualitas[x] <= 70:
    Bagus = 1
    nilaiKain[2] = Bagus
  elif kualitas[x] > 70 and kualitas[x] < 75:
    Bagus = (75 - kualitas[x])/5
    SangatBagus = (kualitas[x] - 70)/5
    nilaiKain[2] = Bagus
    nilaiKain[3] = SangatBagus
  elif kualitas[x] >= 75:
    SangatBagus = 1
    nilaiKain[3] = SangatBagus

  if harga[x] <= 3:
    Murah = 1
    nilaiHarga[0] =Murah
  elif harga[x] > 3 and harga[x] < 5:
    Murah = (5 -  harga[x])/2
    Sedang = (harga[x] - 3)/2
    nilaiHarga[0] = Murah
    nilaiHarga[1] = Sedang
  elif harga[x] >= 5 and harga[x] <= 6:
    Sedang = 1
    nilaiHarga[1] = Sedang
  elif harga[x] > 6 and harga[x] < 8:
    Sedang = (8 - harga[x])/2
    Mahal = (harga[x] - 6)/2
    nilaiHarga[1] = Sedang
    nilaiHarga[2] = Mahal
  elif harga[x] >=8 and harga[x] <= 10:
    Mahal = 1
    nilaiHarga[2] = Mahal
    
#Inference
#Tidak Direkomendasikan
  TR = []
  if nilaiHarga[0] == Murah and nilaiKain[0] == SangatJelek:
    TR.append(min(nilaiKain[0],nilaiHarga[0]))
  if nilaiHarga[1] == Sedang and nilaiKain[0] == SangatJelek:
    TR.append(min(nilaiKain[0],nilaiHarga[1]))
  if nilaiHarga[2] == Mahal and nilaiKain[0] == SangatJelek:
    TR.append(min(nilaiKain[0],nilaiHarga[2]))
  if nilaiHarga[1] == Sedang and nilaiKain[1] == Jelek:
    TR.append(min(nilaiKain[1],nilaiHarga[1]))
  if nilaiHarga[2] == Mahal and nilaiKain[1] == Jelek:
    TR.append(min(nilaiKain[1],nilaiHarga[2]))
  if nilaiHarga[2] == Mahal and nilaiKain[2] == Bagus:
    TR.append(min(nilaiKain[2], nilaiHarga[2]))
  nilaiTR = max(TR)
#Rekomendasi
  R = []
  if nilaiHarga[0] == Murah and nilaiKain[1] == Jelek:
    R.append(min(nilaiKain[1], nilaiHarga[0]))
  if nilaiHarga[0] == Murah and nilaiKain[2] == Bagus:
    R.append(min(nilaiKain[2],nilaiHarga[0]))
  if nilaiHarga[1] == Sedang and nilaiKain[2] == Bagus:
    R.append(min(nilaiKain[2], nilaiHarga[1]))
  if nilaiHarga[2] == Mahal and nilaiKain[3] == SangatBagus:
    R.append(min(nilaiKain[3],nilaiHarga[2]))
  nilaiR = max(R)
#Sangat Rekomendasi
  SR = []
  if nilaiHarga[0] == Murah and nilaiKain[3] == SangatBagus:
    SR.append(min(nilaiKain[3],nilaiHarga[0]))
  if nilaiHarga[1] == Sedang and nilaiKain[3] == SangatBagus:
    SR.append(min(nilaiKain[3],nilaiHarga[1]))
  nilaiSR = max(SR)

#Deffuzyfication
  pembagi = nilaiTR + nilaiR + nilaiSR
  z = ((nilaiTR*25)+(nilaiR*50)+(nilaiSR *100))/pembagi
  print("ID              :",id[x])
  print("Kualitas        :",kualitas[x])
  print("Harga           :",harga[x])
  print("Nilai Kelayakan :", z)
  print("==============================")
  y.append([id[x],kualitas[x], harga[x],z])
#Mengurutkan Hasil
hasil = sorted(y,key = lambda x:x[3],reverse=True)
#Output 5 Supplier Terbaik
hasil_terbaik = hasil[:5]
df = pd.DataFrame(hasil_terbaik, columns=["ID", "Kualitas", "Harga", "Kelayakan"])
(df)
```
